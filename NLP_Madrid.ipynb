## Lemmatization and Stemming
Both processes aim to reduce words to their base or root form, but they do so in different ways.

*Stemming*: Often uses a heuristic process to chop off the ends of words in the hope of achieving the goal correctly most of the time.

*Lemmatization*: Usually involves a more sophisticated method to obtain the base or dictionary form of a word, known as the lemma.

Exercise: Demonstrate stemming and lemmatization on a list of words.

```python
nltk.download('wordnet')  # Download necessary datasets
from nltk.stem import WordNetLemmatizer, PorterStemmer

lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer()

words = ["running", "jumps", "easily", "worried"]

# Lemmatization
lemmatized = [lemmatizer.lemmatize(word) for word in words]
print("Lemmatized Words:", lemmatized)

# Stemming
stemmed = [stemmer.stem(word) for word in words]
print("Stemmed Words:", stemmed)
```

# Text Classification
Text Classification

Theory: Text classification is the process of categorizing text into a set of predefined classes or groups. It can be used for a variety of tasks such as spam detection, genre classification, and more. Machine learning models, typically classifiers, are trained on a dataset with pre-labeled categories.


## Text Classification with TF-IDF Vectorizer and Naive Bayes Classifier

Text classification involves assigning categories to text documents, and it's a common task in Natural Language Processing (NLP). Two key components in this process are the **TF-IDF Vectorizer** and the **Naive Bayes Classifier**. Here's how they work:

### TF-IDF Vectorizer

The **Term Frequency-Inverse Document Frequency (TF-IDF)** Vectorizer transforms text into a numerical format that machine learning algorithms can work with. It does so by converting the text into a matrix of TF-IDF features.

- **Term Frequency (TF)**: Measures how frequently a term occurs in a document. It's the number of times a term appears in a document divided by the total number of terms in the document.

- **Inverse Document Frequency (IDF)**: Measures the importance of a term across a set of documents. It's calculated as the logarithm of the number of documents divided by the number of documents containing the term.

The TF-IDF value increases proportionally to the number of times a word appears in the document but is offset by the frequency of the word across the corpus. This helps to adjust for the fact that some words appear more frequently in general.

### Naive Bayes Classifier

The **Naive Bayes Classifier** is a probabilistic classifier based on applying Bayes' theorem with strong (naive) independence assumptions between the features. It's particularly suited for text classification because of its simplicity and effectiveness.

- When applied to text classification, the classifier calculates the probability of a document belonging to a certain class based on the TF-IDF features. The class with the highest probability is chosen as the prediction.

- Despite its simplicity, Naive Bayes can outperform more complex classifiers, especially in cases with limited training data, due to its assumption of independence between features.

### Implementation Example in Python

This example demonstrates text classification using a simple Naive Bayes classifier along with TF-IDF for feature extraction.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
from sklearn.datasets import fetch_20newsgroups
```

TfidfVectorizer:

Transforms text to feature vectors that can be used as input to estimator. It measures the originality of a word by comparing its frequency in a document with its frequency in other documents.

MultinomialNB:

Implements the naive Bayes algorithm for multinomially distributed data, and is suitable for text classification.

make_pipeline:

Simplifies the process of constructing a processing pipeline by combining the TfidfVectorizer and MultinomialNB into a single step.

train_test_split:

Splits the dataset into a training set and a testing set, allowing the model to learn from one subset and be tested on another.

fetch_20newsgroups:

Loads the 20 newsgroups dataset, a collection of approximately 20,000 newsgroup documents, categorized into 20 different newsgroups.

Loading the Dataset
The dataset is filtered to only include documents from the four specified categories

```python

# Load dataset
categories = ['talk.religion.misc', 'soc.religion.christian', 'sci.space', 'comp.graphics']
data = fetch_20newsgroups(subset='train', categories=categories)
```

The dataset is divided into training and testing sets. X_train and X_test contain the text documents, while y_train and y_test contain the corresponding category labels. test_size=0.25 means 25% of the data is used for testing.

```python
# Preparing data
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.25, random_state=42)
```

This pipeline performs two main tasks: first, it converts the text documents into TF-IDF feature vectors; then, it applies the Naive Bayes classifier to the vectors.

```python
# Model pipeline
model = make_pipeline(TfidfVectorizer(), MultinomialNB())
```

The model learns to classify documents into categories using the training data.

```python
# Train the model
model.fit(X_train, y_train)
```

given a new document, the model predicts which category it belongs to.

```python

# Predictions
predicted_categories = model.predict(["Jesus Christ is the central figure of Christianity."])

# Output the corresponding category
print(data.target_names[predicted_categories[0]])
```

The predicted category index is used to find the corresponding category name, which is then printed. In this case, the model should ideally predict 'soc.religion.christian' for the given text.
This example demonstrates how machine learning can be applied to text classification, automating the process of categorizing documents into predefined groups based on their content.

## Universal Sentence Encoder (USE) for text embedding

The Universal Sentence Encoder (USE) is a machine learning model designed to convert text into high-dimensional vector embeddings. These embeddings are capable of capturing the semantic meaning of sentences, which makes them highly useful for a wide range of Natural Language Processing (NLP) tasks, including text classification, semantic similarity measurement, and clustering.

```python
import requests
import tensorflow_hub as hub

# Fetch the text from Project Gutenberg
url = "https://www.gutenberg.org/files/28054/28054-0.txt"
response = requests.get(url)
if response.status_code == 200:
    text = response.text[min_chars:max_chars]  # Extract specified character range
else:
    print("Failed to retrieve text. Status code:", response.status_code)

# Load the Universal Sentence Encoder
embed = hub.load("https://tfhub.dev/google/universal-sentence-encoder/4")

# Generate embedding for the extracted text segment
text_embedding = embed([text])

# Example: Displaying the shape of the embedding
print(text_embedding.shape)
```

```python
text_embedding
```

Each of the 512 dimensions in the embedding vector represents a semantic feature of the input text, learned by the USE from a vast corpus of text during its training. These features are not easily interpretable by humans, as they are high-dimensional representations of the text's meaning, capturing various aspects of semantic information such as topics, sentiment, and syntax.

The embedding can be used in numerous NLP tasks, including but not limited to:

Semantic text similarity:

Comparing embeddings to determine how similar two pieces of text are in terms of their meaning.

Text classification:

Using embeddings as input features for machine learning models to classify texts into predefined categories.

Clustering:

grouping texts into clusters based on the similarity of their embeddings, which can help in discovering latent topics or themes within a large collection of texts.
Remember, while the embedding captures the semantic essence of the text, interpreting the individual dimensions of the embedding vector directly is not straightforward. Instead, the utility of these embeddings lies in their ability to serve as input for downstream tasks where semantic understanding of text is crucial.

The given code snippet demonstrates how to generate text embeddings using the Universal Sentence Encoder from TensorFlow Hub, reduce the dimensionality of these embeddings with PCA (Principal Component Analysis) to two dimensions, and then plot the results for visualization.

```python
import requests
import tensorflow_hub as hub
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Example texts
texts = [
    "The quick brown fox jumps over the lazy dog.",
    "I love watching science fiction movies and reading books about space.",
    "Religious texts often explore the meaning of existence and morality.",
    "Computer graphics have revolutionized the way we visualize data.",
    "Quantum computing is a breakthrough in processing information.","A nimble deer leaps over a tranquil stream in the forest.",
"Exploring alien worlds in novels fuels my imagination about the universe.",
"Philosophical discussions often delve into concepts of free will and destiny.",
"Advancements in virtual reality enhance our immersion into digital worlds.",
"The potential of artificial intelligence extends far beyond current technologies.",

]

# Load the Universal Sentence Encoder
embed = hub.load("https://tfhub.dev/google/universal-sentence-encoder/4")

# Generate embeddings for each text
embeddings = embed(texts)

# Reduce dimensions with PCA
pca = PCA(n_components=2)
reduced_embeddings = pca.fit_transform(embeddings)

# Plot
plt.figure(figsize=(10, 8))
for i, text in enumerate(texts):
    plt.scatter(reduced_embeddings[i, 0], reduced_embeddings[i, 1])
    plt.text(reduced_embeddings[i, 0]+0.01, reduced_embeddings[i, 1]+0.01, text, fontsize=9)
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('Text Embeddings Visualized with PCA')
plt.show()

```

## embeding of random lines of classical books

```python
import requests

# URLs de los libros en Project Gutenberg
urls = {
    "Don Quijote": "https://www.gutenberg.org/files/996/996-0.txt",
    "El Conde de Montecristo": "https://www.gutenberg.org/files/1184/1184-0.txt",
    "Dracula": "https://www.gutenberg.org/files/345/345-0.txt"
}

# Funcin para descargar y seleccionar 100 frases de un libro
def fetch_and_select_sentences(url, num_sentences=100):
    response = requests.get(url)
    if response.status_code == 200:
        text = response.text
        sentences = text.split('.')  # Dividir el texto en frases
        selected_sentences = sentences[:num_sentences]  # Seleccionar las primeras 100 frases
        return selected_sentences
    else:
        return []

# Descargar y seleccionar frases para cada libro
# Nota: Este paso es demostrativo. La ejecucin real requiere ajustes debido al lmite de ejecucin y a las prcticas de uso de la API.
sentences_per_book = {}
for book, url in urls.items():
    sentences_per_book[book] = fetch_and_select_sentences(url)

# Mostrar la cantidad de frases seleccionadas para verificar
{book: len(sentences) for book, sentences in sentences_per_book.items()}
```

```python
import requests
import tensorflow_hub as hub
import numpy as np
from sklearn.decomposition import PCA
import plotly.graph_objects as go
from umap import UMAP
import random
import re

# Function to fetch text from URL
def fetch_text(url):
    response = requests.get(url)
    return response.text if response.status_code == 200 else None

# URLs for the books
book_urls = {
    "Don Quixote": "https://www.gutenberg.org/files/996/996-0.txt",
    "The Count of Monte Cristo": "https://www.gutenberg.org/files/1184/1184-0.txt",
    "Dracula": "https://www.gutenberg.org/files/345/345-0.txt",
    "Pride and Prejudice": "https://www.gutenberg.org/files/1342/1342-0.txt",
}

# Load the Universal Sentence Encoder
embed = hub.load("https://tfhub.dev/google/universal-sentence-encoder/4")

book_sentences = {}
all_sentences = []  # To store all sentences
labels = []  # To store book labels for each sentence

for book, url in book_urls.items():
    text = fetch_text(url)
    if text:
        # Split text into sentences
        sentences = re.split(r'\. |\?|!', text)
        sentences = [sentence.strip() for sentence in sentences if sentence.strip()]
        if len(sentences) > 100:
            selected_sentences = random.sample(sentences, 100)
        else:
            selected_sentences = sentences
        book_sentences[book] = selected_sentences
        all_sentences.extend(selected_sentences)
        labels.extend([book] * len(selected_sentences))

# Generate embeddings for all sentences
all_embeddings = embed(all_sentences)

# PCA for dimensionality reduction
pca = PCA(n_components=3)
pca_embeddings = pca.fit_transform(all_embeddings)

# UMAP for dimensionality reduction
umap = UMAP(n_components=3, n_neighbors=10)
umap_embeddings = umap.fit_transform(all_embeddings.numpy())

# Colors for each book
colors = {
    'Don Quixote': 'red',
    'The Count of Monte Cristo': 'blue',
    'Dracula': 'green',
    'Pride and Prejudice': 'purple',
}

# Plot PCA 3D
fig_pca = go.Figure()
for book in book_urls.keys():
    idx = [i for i, label in enumerate(labels) if label == book]
    fig_pca.add_trace(go.Scatter3d(
        x=pca_embeddings[idx, 0], y=pca_embeddings[idx, 1], z=pca_embeddings[idx, 2],
        mode='markers',
        marker=dict(size=5, color=colors[book]),
        text=[all_sentences[i] for i in idx],
        name=book
    ))
fig_pca.update_layout(title="PCA Dimensionality Reduction", margin=dict(l=0, r=0, b=0, t=0), hovermode='closest')
fig_pca.show()

# Plot UMAP 3D
fig_umap = go.Figure()
for book in book_urls.keys():
    idx = [i for i, label in enumerate(labels) if label == book]
    fig_umap.add_trace(go.Scatter3d(
        x=umap_embeddings[idx, 0], y=umap_embeddings[idx, 1], z=umap_embeddings[idx, 2],
        mode='markers',
        marker=dict(size=5, color=colors[book]),
        text=[all_sentences[i] for i in idx],
        name=book
    ))
fig_umap.update_layout(title="UMAP Dimensionality Reduction", margin=dict(l=0, r=0, b=0, t=0), hovermode='closest')
fig_umap.show()
```
